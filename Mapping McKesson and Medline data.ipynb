```python
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
import pandas as pd
import numpy as np
```

```python
#Get the list of query strings
# Recursively input each query string to function
# get the index value as output and collect them in a Dataframe along with respective query string
```

```python
def match_output(query_str,full_list):
    temp=0
    i_temp=0
    for i in range(len(full_list)):
        ratio=fuzz.token_set_ratio(query_str,full_list[i])
        if ratio>90 :
            i_temp=i
            break
        elif ratio>temp:
            temp=ratio
            i_temp=i
            
    return i_temp
```
