```python
col_names=['Secodary_key', 'Dist_part_num_x', 'Distributor_x', 'Manufacturer_x',
       'Mfg_part_num_x', 'Prod_desc_x', 'Final_cost_x','mapped_key']
```

```python
result=pd.DataFrame(columns=col_names)
```

```python
HS_to_map.head()
```

```python
for idx, row in HS_to_map.iterrows():
    spn_to_map=row['Dist_part_num_x']
    
    if (spn_to_map not in [np.nan]):
        spn_to_map=spn_to_map.strip()
        op=HS_DF_PK_spn[HS_DF_PK_spn['Item #']==spn_to_map]
        
        if (len(op)>0):
            op=op[:1]
            result=result.append({'Secodary_key':row['Secodary_key'], 
                                  'Dist_part_num_x':row['Dist_part_num_x'],
                                  'Distributor_x':row['Distributor_x'],
                                  'Manufacturer_x':row['Manufacturer_x'],
                                  'Mfg_part_num_x':row['Mfg_part_num_x'],
                                  'Prod_desc_x':row['Prod_desc_x'],
                                  'Final_cost_x':row['Final_cost_x'],
                                  'mapped_key':op['Primary_Key'].get_values()[0]},ignore_index=True)
            print(spn_to_map)
    
```
