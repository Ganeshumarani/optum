```python
invoice_file=pd.read_excel("MountainWest_Invoice_010118_310519.xlsx")
```

```python
po_file=pd.read_csv("MountainWest_PO_010118_310519_BothFiles.csv")
```

```python
invoice_file.columns
```

```python
# Take each 'InvoiceID' in invoice file and filter by that value in invoice file.
# for each 'InvoiceDesc' in the above filtered result find the best match of product description for that Invoice ID in PO file
# Extract the PK of the matched PD in po file and append it to a dataframe along with all the information of the line in invoice file
```

```python
list_of_inv_id=list(set(invoice_file['InvoiceID'].tolist()))
```

```python
col_names=['Primary_Key', 'Key_BDPaaS', 'SETID', 'SupplierID', 'SupplierName',
       'Commodity', 'Certification', 'AccountingDate', 'BU', 'OU', 'OU_Desc',
       'Account', 'Account_Desc', 'Dept', 'InvoiceID', 'Approver', 'Requestor',
       'Last_Approver', 'Amount', 'InvoiceDesc', 'LocationCode',
       'Active_Status', 'SubCategory', 'Address1', 'City', 'State', 'Postal',
       'L1', 'L1 NAME', 'L2', 'L2 NAME', 'L3', 'L3 NAME', 'L4', 'L4 Name',
       'CDO','PO_PK']
```

```python
op_res=pd.DataFrame(columns=col_names)
```

```python
op_res
```

```python
for i in range(len(list_of_inv_id)):
    print(list_of_inv_id[i])
    inv_file_subset=invoice_file[invoice_file.InvoiceID==list_of_inv_id[i]]
    po_file_subset=po_file[po_file['Invoice number']==list_of_inv_id[i]]

    if (len(po_file_subset)>0):
        full_list=po_file_subset['Product Description'].tolist()
        for idx,row in inv_file_subset.iterrows():
            desc=row['InvoiceDesc']

            if desc not in [np.nan]:
                if process.extractOne(desc,full_list,score_cutoff=70) is not None:
                    matched_str,ratio=process.extractOne(desc,full_list,score_cutoff=70)
                    pk=po_file_subset[po_file_subset['Product Description']==matched_str]['Primary_Key'].get_values()[0]
                    op_res=op_res.append({'Primary_Key':row['Primary_Key'],
                                         'Key_BDPaaS':row['Key_BDPaaS'],
                                         'SETID':row['SETID'],
                                         'SupplierID':row['SupplierID'],
                                         'SupplierName':row['SupplierName'],
                                         'Commodity':row['Commodity'],
                                         'Certification':row['Certification'],
                                          'AccountingDate':row['AccountingDate'],
                                          'BU':row['BU'],
                                          'OU':row['OU'],
                                          'OU_Desc':row['OU_Desc'],
                                          'Account':row['Account'],
                                          'Account_Desc':row['Account_Desc'],
                                          'Dept':row['Dept'],
                                          'InvoiceID':row['InvoiceID'],
                                          'Approver':row['Approver'],
                                          'Requestor':row['Requestor'],
                                          'Last_Approver':row['Last_Approver'],
                                          'Amount':row['Amount'],
                                          'InvoiceDesc':row['InvoiceDesc'],
                                          'LocationCode':row['LocationCode'],
                                          'Active_Status':row['Active_Status'],
                                          'SubCategory':row['SubCategory'],
                                          'Address1':row['Address1'],
                                          'City':row['City'],
                                          'State':row['State'], 
                                          'Postal':row['Postal'],
                                          'L1':row['L1'],
                                          'L1 NAME':row['L1 NAME'],
                                          'L2':row['L2'], 
                                          'L2 NAME':row['L2 NAME'], 
                                          'L3':row['L3'], 
                                          'L3 NAME':row['L3 NAME'], 
                                          'L4':row['L4'], 
                                          'L4 Name':row['L4 Name'],
                                          'CDO':row['CDO'], 
                                          'PO_PK':pk},ignore_index=True)

```

```python
op_res.to_csv("MountainWest_inv_po_mapped.csv")
```
